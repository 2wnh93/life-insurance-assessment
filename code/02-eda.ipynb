### Response feature
As this is a classification problem, the first thing I will look at is the target variable (ie. the classes that we will be predicting). If there is are imbalanced classes, I might want to balance the classes first, prior to modelling. Alternatively, I might want to consider using `F1 score` which is a classification metric that considers imbalanced classes. 

```python
# plot countplot of response variable 
ax = sns.countplot(x = 'Response', data = train)
plt.title("Distribution of Response Feature", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
plt.xlabel("Response", fontsize = 14)

for p in ax.patches:
    x = p.get_bbox().get_points()[:,0]
    y = p.get_bbox().get_points()[1,1]
    ax.annotate(f"{round(100*y/len(train),1)}%", (x.mean(),y),
                ha = 'center', va = 'bottom')
#save file
plt.savefig("../assets/distribution-response.png")
plt.show()
```

There is a significantly higher proportion of Class 8 (32.8%) compared to other classes. 

Total applicants or prospects from this class is the equivalent to about 3 other classes combined. It might be worth to focus on just this one class. For instance, it may not be worth looking at characteristics of applicants beloning to Class 3 since these are just 1.7% of the total number of applicants (assuming this sample is representative of the entire population). Another way to appraoch the modelling here is to do a binary classification instead where one class is risk class of 8 and the other will comprise other risk classes.

With these imbalanced classes, I will balance these classes first, prior to modelling.
