### Response feature
As this is a classification problem, the first thing I will look at is the target variable (ie. the classes that we will be predicting). If there is are imbalanced classes, I might want to balance the classes first, prior to modelling. Alternatively, I might want to consider using `F1 score` which is a classification metric that considers imbalanced classes. 

```python
# plot countplot of response variable 
ax = sns.countplot(x = 'Response', data = train)
plt.title("Distribution of Response Feature", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
plt.xlabel("Response", fontsize = 14)

for p in ax.patches:
    x = p.get_bbox().get_points()[:,0]
    y = p.get_bbox().get_points()[1,1]
    ax.annotate(f"{round(100*y/len(train),1)}%", (x.mean(),y),
                ha = 'center', va = 'bottom')
#save file
plt.savefig("../assets/distribution-response.png")
plt.show()
```

There is a significantly higher proportion of Class 8 (32.8%) compared to other classes. 

Total applicants or prospects from this class is the equivalent to about 3 other classes combined. It might be worth to focus on just this one class. For instance, it may not be worth looking at characteristics of applicants beloning to Class 3 since these are just 1.7% of the total number of applicants (assuming this sample is representative of the entire population). Another way to appraoch the modelling here is to do a binary classification instead where one class is risk class of 8 and the other will comprise other risk classes.

Since there are imbalanced classes, I will balance these classes first, prior to modelling. 

### Interaction with Response
Here I explore how different sets of variables interact with the risk classes. I first create functions to explore the interactions for categorical and continuous variables respectively. 

```python
# function that plots violinplots of categorical variables against Response
def cat_var_interactions_with_response(list_of_x, title):
    '''input: list of variables to view correlation with Response, title of plot, and xlabel
    output: scatterplot showing variables and response'''
    for i in list_of_x:
        plt.figure(figsize = (15,6))
        sns.violinplot(x=i, y='Response', data=train)
        plt.title(title)
        plt.xlabel(i)
        plt.ylabel("Response")
        plt.show()    
```

```python
# function that plots scatter points of continuous variables against Response
def cont_var_interactions_with_response(list_of_x, title):
    '''input: list of variables to view correlation with Response, title of plot, and xlabel
    output: scatterplot showing variables and response'''
    for i in list_of_x:
        plt.figure(figsize = (15,6))
        sns.scatterplot(x=i, y='Response', data=train)
        plt.title(title)
        plt.xlabel(i)
        plt.ylabel("Response")
        plt.show()    
```

#### <ins>Product info with Response</ins>

```python
# apply function to view interactions between product info types and response
pdt_info = ['Product_Info_1','Product_Info_3','Product_Info_4','Product_Info_5','Product_Info_6','Product_Info_7']
cat_var_interactions_with_response(pdt_info, "Product Info Against Response")
```

**From the violinplots, we have the following observations :**

1. For `Product_Info_1` where values could be either 1 or 2, both show a similar interaction to risk class. Majority of insurance applicants fall under Risk class 6 and above, or 2 and below. This is the same for `Product_Info_5` and `Product_Info_6`.


2. For `Product_Info_7`, insurance applicants with value 2 are all in risk class 8, while those with other values (1 or 3) follow the same interaction as `Product_Info_1, 5 and 6`.


3. `Product_Info_3` and `Product_Info_4` have a range of values with indeterminate interactions with the different risk classes. 

**Actions to consider :**

Since there is a general trend abovementioned, a way to reduce the complexity of the model (by reducing the number of features) is to drop `Product_Info_1, 5, 6` and change `Product_Info_7` into a binary where values 1 and 3 are lumped together into a class, and values 2 are put into another class. Leave `Product_Info_3 and 4` as is. 

#### <ins>Insurance Age, Height, Weight and BMI against Response</ins>

```python
# apply function or continuous variables against Response 
cont_var = ['Ins_Age','Ht','Wt','BMI']
cont_var_interactions_with_response(cont_var, "Age, Height, Weight, BMI against Response")
```

**From the scatterplots, we have the following observations :**

1. For `Age`, all applicants are above a particular age which is normalized to about 0.7. Applicants which are older, surprisingly fall under the risk class 1 category or class 7/6. 


2. `Ht` of insurance applicants fall within range of 0.4 to 1.0 but there is no visible pattern as to its interaction with the different risk classes. There are some outliers, which we should remove, specifically, those that *fall below 0.3*.


3. `Wt` has the largest range for risk class 1. Apart from this observation, there isn't anything stark about weight distribution among the rest of the risk classes. There are some outliers which should be removed as well, specifically those values that fall *less than 0.02* and *more than 0.9*.


4. `BMI` across all classes range from about 0.2 onwards, with the higher BMI appearing mostly in risk classes 1 and 2. Outliers which fall *less than 0.1* should be removed. 

**Actions to take :**

Drop outliers.

```python
# drop outliers abovementioned
print(f"Train shape before removing outliers : {train.shape}")
train = train.loc[~((train['Ht']<0.3)|(train['Wt']<0.02)|(train['Wt']>0.9)|(train['BMI']<0.1)),:]
print(f"Train shape after removing outliers : {train.shape}")
```
